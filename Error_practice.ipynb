```python
# Import dependencies
import matplotlib.pyplot as plt
import statistics as stat
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis
stdev = stat.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Using OOP 
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# my own practice in showing data with a low standard dev
sample_y_axis = [4,5,6,7,4,5,6,7,4,5,6,5]
sample_dev = stat.stdev(sample_y_axis)
sample_dev
print(stat.mean(sample_y_axis))
print(sample_dev)
```

```python
plt.errorbar(x_axis, sample_y_axis, yerr=sample_dev)
```

```python
# another practice but with high standard dev
wide_sample_y = [1,4,76,50,20,120,33,78,22,70,150,30]
wide_dev = stat.stdev(wide_sample_y)
print(wide_dev)
print(stat.mean(wide_sample_y))
```

```python
plt.errorbar(x_axis, wide_sample_y, yerr=wide_dev)
```
