```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot - this is using the MATLAB method
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - this is the OOP method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create a quick graph with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = '*', color='blue', linewidth=2, label='Boston')

# Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title('PyBer Fare by Month')

# Add a grid
plt.grid()

# Add a legend
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Skill drill, matlab way
fig, ax = plt.subplots()
ax.plot(y_axis, x_axis, 'o', color='red', label='Chicago')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend()
ax.invert_yaxis()
ax.set_xlim(0,50)
```

```python
# OOP scatter plot
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Skill drill
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c = 'skyblue', s = [i * 5 for i in y_axis], label = 'Boston', alpha = .8, edgecolors='black')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend()
ax.invert_yaxis()
ax.set_xlim(0, 50)
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Pie data
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assgin 12 colors, one for each month
colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen', 'greenyellow', 'yellow', 'orange', 'gold', 'indianred', 'tomato', 'mistyrose']
explode_values = [0,0,0,0,0,0,0.2,0,0,0,0,0]
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')
plt.show()
```
